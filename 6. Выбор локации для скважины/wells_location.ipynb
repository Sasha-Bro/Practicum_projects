# Описание проекта

Допустим, вы работаете в добывающей компании «ГлавРосГосНефть». Нужно решить, где бурить новую скважину.

Вам предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов. Постройте модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Проанализируйте возможную прибыль и риски техникой *Bootstrap.*

Шаги для выбора локации:

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

**Условия задачи:**

- Для обучения модели подходит только линейная регрессия (остальные — недостаточно предсказуемые).

- При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.

- Бюджет на разработку скважин в регионе — 10 млрд рублей.

- При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.

- После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.

Данные синтетические: детали контрактов и характеристики месторождений не разглашаются.

### Создание функций

Создадим функции, которые нам потребуются в работе.

```python
# Функция для предварительного изучения 
# общей информации о данных.  
def data_research(dataset):
    
    display(dataset.head())
    display(dataset.tail())
    print()
    
    
    display(HTML('<h3>Общая информация о данных</h3><br>'))
    print(dataset.info())
    
    
    display(HTML('<h3>Наименования столбцов</h3><br>'))
    print(dataset.columns)
 


    display(HTML('<br><h3>Количество пропущенных значений</h3><br>'))
    
    if dataset.isna().sum().any():
    
        
        print(dataset.isna().sum())
    
        display(HTML('<br><h3>Доля пропущенных значений</h3><br>'))
        display(HTML(
            f'<b>Во всем датафрейме: </b>'\
            f'{round((dataset.isna().sum().sum() / (dataset.shape[0] * dataset.shape[1]) * 100), 1)}%<br>')
            )
               
        display(HTML('<b>В каждом столбце: </b>'))
        display(pd.DataFrame(dataset.isna().mean()*100)\
                .style.set_precision(1).background_gradient('coolwarm'))
    
    else:
        print('Пропущенные значения отсутствуют')
    
    
    display(HTML('<br><h3>Количество дубликатов в датафрейме</h3><br>'))
    
    print(f'Всего: {dataset.duplicated().sum()} '
          f'({(dataset.duplicated().sum() / dataset.shape[0]):.2%} '\
          f'от общего размера)')

    
    
    if len(dataset.select_dtypes(include=[np.number]).columns) != 0:

        display(HTML('<br><h3>Описание данных</h3>'))
        display(dataset.describe())
        
        display(HTML('<br><h3>Гистограммы для всех числовых столбцов</h3><br>'))
        axes = dataset.hist(figsize=(10, 10), xlabelsize=10, ylabelsize=10)

        # Добавляем подписи осей
        for ax in axes.flatten():
            ax.set_xlabel(ax.get_title(), fontsize=12)
            ax.set_ylabel('Количество записей', fontsize=12)

        plt.tight_layout()
        plt.show()
   
    else:
        pass
```

```python
# Функция для построения графиков для количественных признаков.
def int_plot(data, column):
    # Создаем фигуру с двумя подграфиками (2 строки, 1 столбец).
    fig, axs = plt.subplots(2, 1, figsize=(15, 10))
    
    # График распределения.
    sns.histplot(data=data, 
                 x=column, 
                 ax=axs[0], 
                 kde=True,
                 bins=20)
    axs[0].grid(True)
    axs[0].set_xlabel(column, fontsize=12)
    axs[0].set_ylabel('Количество скважин', fontsize=12)
    
    # Диаграмма размаха (боксплот).
    sns.boxplot(data=data, x=column, ax=axs[1])
    axs[1].grid(True)
    axs[1].set_xlabel(column, fontsize=12)
    
    # Общий заголовок для графиков.
    plt.suptitle(f'График распределения и '\
                 f'диаграмма размаха значений столбца "{column}"', 
                 fontsize=14)
    plt.tight_layout()
    plt.show()
```

###  Корреляционный анализ

```python
for data_name in all_data:
    data = eval(data_name)
    
    display(HTML(f'<br><h3>Матрица корреляции признаков `{data_name}`</h3>'))
    display(data.phik_matrix(interval_cols=data.columns)\
                .style.background_gradient('cividis'))
```

В первом регионе высокой корреляцией обладают `'f0' и 'f1'` и умеренной `'f2' и 'product'`. В третьем регионе имеется лишь умеренная связь у `'f2' и 'product'`.

Во втором регионе заметную связь имеют `'f0' и 'product'`, а также черезвычайно высокую корреляцию имеют `'f2' и 'product'`, практически равной 1. 

Проверим признаки при помощи коэффициента инфляции дисперсии (VIF).

```python
for data_name in all_data:
    data = eval(data_name)
    
    display(HTML(f'<br><h3>Коэффициента инфляции дисперсии (VIF) '
                 f'`{data_name}`</h3>'))
    
    vif_data = pd.DataFrame()
    vif_data['feature'] = data.columns
    vif_data['VIF'] = [variance_inflation_factor(data.values, i) \
                       for i in range(data.shape[1])]

    display(vif_data.style.background_gradient('YlGnBu'))
```

## Обучение и проверка модели

Создадим функцию, в которой будет одновременно проведена подготовка данных, обучение модели, предсказание целевого признака и расчет метрик качества модели.

```python
def model_results(data):
    
    # Разделение данных на признаки и целевую переменную.
    X = data.drop(['product'], axis=1)
    y = data['product']

    # Разделение данных на обучающую и тестовую выборки.
    X_train, X_test, y_train, y_test = train_test_split(
        X, 
        y, 
        test_size=0.25, 
        random_state=RANDOM_STATE
    )
    
    # Создание пайплайна.
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', LinearRegression())
    ])
    
    # Обучение модели.
    pipeline.fit(X_train, y_train)
    
    # Предсказания на тестовой выборке.
    y_pred = pipeline.predict(X_test)
    
    # Вычисление метрик.
    rmse = round(np.sqrt(mean_squared_error(y_test, y_pred)), 2)
    mae = round(mean_absolute_error(y_test, y_pred), 2)
    r2 = round(r2_score(y_test, y_pred), 2)
    
    # Создание словаря с метриками.
    model_metrics = {
        'rmse': rmse,
        'mae': mae,
        'r2': r2,
        'true_mean': round(y_test.mean(), 2),
        'pred_mean': round(y_pred.mean(), 2)
    }
    
    true_pred_df = pd.DataFrame({
        'true_product': y_test,
        'pred_product': y_pred
    })
        
    return model_metrics, true_pred_df
```

Выведем все метрики по регионам в таблицу.

```python
# Создание пустого датафрейма для хранения результатов.
results_df = pd.DataFrame()

for data_name in all_data:
    data = eval(data_name)
    # Преобразуем словарь во временный датафрейм.
    df = pd.DataFrame([model_results(data)[0]], index=[data_name])

    # Собираем в единый датафрейм.
    results_df = pd.concat([results_df, df])

display(results_df)
```

У регионов 1 и 3  **RMSE**  и  **MAE**  достаточно высокие, что указывает на значительные ошибки предсказания. Низкий коэффициент детерминации R² свидетельствует о слабой объяснительной способности модели.

У региона  2 **RMSE**  и  **MAE**  очень низкие, что указывает на высокую точность предсказаний. Коэффициент детерминации R² равен 1, что свидетельствует о практически идеальной модели - это связано с очень высокой корреляцией целевого признака с `'f2'`. 

По всем регионам средние значения истинных и предсказанных значений практически идентичны.

## Расчёт прибыли и рисков

Напишем функцию расчета прибыли по 200 прогнозируемо лучшим скважинам.

```python
def wells_profit(data, column):
    best_wells = \
        data.sort_values(by='pred_product',ascending=False)[:BEST_WELLS_NUM]
    profit = best_wells[column].sum() * COST_1000_BARREL - BUDGET
    return profit
```

Считаем прибыль по каждому региону.

```python
# Создаем новые столбцы в таблице для сохранения результатов.
results_df[['true_total_profit', 'pred_total_profit']] = 0

for i, data_name in enumerate(all_results):
    data = eval(data_name)
    
    for n, m in [(7, 0), (8, 1)]:
        results_df.loc[results_df.index[i], results_df.columns[n]] = \
            round(wells_profit(data, data.columns[m]) / 10**9, 2)
    
display(HTML(f'<h3>Истинная и прогнозируемая прибыль '
             f'по 200 предсказанным лучшим скважинам (млрд. руб.)</h3>'))
display(results_df[results_df.columns[7:]])
```

Прогнозы немного выше истинных значений. По истинным значениям каждый регион принесет прибыль в размере от 24 до 33% от вложенных средств, что является хорошим результатом. Самым прибыльным регионом показывает себя регион 1, наименее прибыльным - регион 2.

Создадим функцию `bootstrap_profit` для применения техники Bootstrap с 1000 выборок, чтобы найти распределение прибыли.

```python
def bootstrap_profit(data, n_samples=1000):
    profits = []
    state = np.random.RandomState(12345)
    for _ in range(n_samples):
        sample = data.sample(n=WELLS_NUM, replace=True, random_state=state)
        profit = wells_profit(sample, 'true_product')
        profits.append(profit)
    return np.array(profits)
```

Создадим функцию для построения гистограмм с вертикальными линиями для средних и доверительного интервала.

```python
def plot_profit_distribution(profits, mean_profit, lower_bound, upper_bound, region_name):
    plt.figure(figsize=(10, 6))
    sns.histplot(profits, bins=30, kde=True, alpha=0.7)
    
    # Вертикальные линии для среднего и доверительного интервала.
    plt.axvline(mean_profit, color='red', linestyle='--', label=f'Средняя прибыль: {mean_profit:.2f} млн руб.')
    plt.axvline(lower_bound, color='green', linestyle='--', label=f'Нижняя граница ДИ: {lower_bound:.2f} млн руб.')
    plt.axvline(upper_bound, color='green', linestyle='--', label=f'Верхняя граница ДИ: {upper_bound:.2f} млн руб.')
    
    # Подписи осей и заголовок.
    plt.xlabel('Прибыль (млн руб.)', fontsize=12)
    plt.ylabel('Частота', fontsize=12)
    plt.title(f'\nРаспределение прибыли для региона {region_name}', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.show()
```

Рассчитаем среднюю прибыль, 95%-й доверительный интервал и риск убытков. Построим графики.

```python
# Создаем новые столбцы в таблице для сохранения результатов.
results_df[['mean_profit', 'lower_bound', 'upper_bound', 'risk_of_loss']] = 0


for i, data_name in enumerate(all_results):
    data = eval(data_name)
    
    # Рассчитываем все нужные величины.
    profits = bootstrap_profit(data) / 10**6
    mean_profit = np.mean(profits)
    lower_bound = np.percentile(profits, 2.5)
    upper_bound = np.percentile(profits, 97.5)
    risk_of_loss = (profits < 0).mean()
        
    # Сохраняем результаты в таблицу.
    results_df.loc[results_df.index[i], 'mean_profit'] = \
        round(mean_profit, 2)
    results_df.loc[results_df.index[i], 'lower_bound'] = \
        round(lower_bound, 2)
    results_df.loc[results_df.index[i], 'upper_bound'] = \
        round(upper_bound, 2)
    results_df.loc[results_df.index[i], 'risk_of_loss'] = risk_of_loss
    
    # Построение гистограммы с вертикальными линиями.
    plot_profit_distribution(profits, mean_profit, lower_bound, upper_bound, all_data[i])
    display(HTML(f'Риск убытков: {risk_of_loss:.1%}'))
```
