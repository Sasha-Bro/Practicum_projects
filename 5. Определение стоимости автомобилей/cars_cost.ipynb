### Создание функций

Создадим функции, которые нам потребуются в работе.

```python
# Функция для отображения списка только 
# категориальных столбцов датафрейма.  
def cat_col(data):
    return data.select_dtypes(exclude=['number', 'datetime64']).columns
```

```python
# Функция для отображения списка только 
# количественных столбцов датафрейма.  
def num_col(data):
    return data.select_dtypes(include=['number', 'datetime64']).columns
```

```python
# Функция для предварительного изучения 
# общей информации о данных.  
def data_research(dataset):
    
    display(dataset.head())
    display(dataset.tail())
    print()
    
    
    display(HTML('<h3>Общая информация о данных</h3><br>'))
    print(dataset.info())
    
    
    display(HTML('<h3>Наименования столбцов</h3><br>'))
    print(dataset.columns)
 

    display(HTML('<br><h3>Количество и доля пропущенных значений</h3><br>'))
    
    if dataset.isna().sum().any():

        display(HTML(
            f'<b>Количество пропусков во всем датафрейме: </b>'\
            f'{dataset.isna().sum().sum()} '\
            f'({round((dataset.isna().sum().sum() / (dataset.shape[0] * dataset.shape[1]) * 100), 1)}%)<br>')
            )
        
        display(HTML(
            f'<b>Записи, имеющие хотя бы один пропуск: </b>'\
            f'{dataset[dataset.columns].isna().any(axis=1).sum()} '\
            f'({round((dataset[dataset.columns].isna().any(axis=1).sum() / dataset.shape[0] * 100), 1)}%)<br>')
            )
               
        display(HTML('<b>Количество пропусков в каждом столбце: </b>'))
        display(pd.DataFrame({
            'count': dataset.isna().sum(),
            '%': dataset.isna().mean() * 100
            }).style.set_precision(1).background_gradient('coolwarm'))
    
    else:
        print('Пропуски отсутствуют')
    
    
    display(HTML('<br><h3>Количество дубликатов в датафрейме</h3><br>'))
    
    print(f'Всего: {dataset.duplicated().sum()} '
          f'({(dataset.duplicated().sum() / dataset.shape[0]):.2%} '\
          f'от общего размера)')
    

    
    display(HTML('<br><h3>Уникальные значения в датафрейме</h3><br>'))    
        
    for column in dataset.columns:
        display(HTML(f'<b>Уникальные значения в столбце "{column}": </b>'\
                     f'{len(dataset[column].unique())}'))
            
        print(dataset.sort_values(by=column)[column].unique())
   
    
    
    if len(num_col(dataset)) != 0:

        display(HTML('<br><h3>Описание данных</h3>'))
        display(dataset.describe())
        
        display(HTML('<br><h3>Гистограммы для всех числовых столбцов</h3><br>'))
        num_plots = len(num_col(dataset))
        num_rows = (num_plots + 1) // 2

        # Создание фигуры и осей
        fig, axes = plt.subplots(num_rows, 2, figsize=(14, 5 * num_rows))

        # Построение гистограмм
        for i, column in enumerate(num_col(dataset)):
            row = i // 2
            col = i % 2
            sns.histplot(data=dataset, x=column, bins=20, ax=axes[row, col])
            axes[row, col].grid(True)
            axes[row, col].set_title(f'Распределение значений признака {column}')
            axes[row, col].set_xlabel(column)
            axes[row, col].set_ylabel('Количество записей')
            axes[row, col].tick_params(axis='x', rotation=90)

        # Скрытие последнего пустого графика, если количество столбцов нечетное
        if num_plots % 2 != 0:
            fig.delaxes(axes[-1, -1])

        # Настройка макета
        plt.tight_layout()
        plt.show()
   
    else:
        pass
```

```python
# Функция для подсчета и отображения отфильтрованных данных.  
def count_n_display(n):
    print('Количество записей:', 
          n.shape[0])
    print(f'Доля таких записей: '
          f'{round(n.shape[0] / data_size * 100, 2)}%')
    display(n.head())
```

```python
# Функция для обучения модели, предсказания результатов
# и получения ее метрик качества.
def model_results(model, model_name, X_train, y_train, X_val, y_val, results):
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time

    start_time = time.time()
    val_pred = model.predict(X_val)
    pred_time = time.time() - start_time

    val_rmse = np.sqrt(mean_squared_error(y_val, val_pred))

    results = results.append(
        {'Model': model_name, 
         'Training Time': training_time, 
         'Prediction Time': pred_time,
         'Validation RMSE': val_rmse}, 
        ignore_index=True)
    
    return model, results
```

### Подготовка данных

Обучим разные модели:
- LinearRegression,
- RandomForestRegressor,

Для каждой модели попробуем подобрать разные гиперпараметры. Но перед этим подготовим данные для обучения моделей.

Делим признаки на входные и целевой.

```python
X = data.drop('price', axis=1)
y = data['price']
```

Разделяем категориальные и количественные признаки.

```python
cat_columns = cat_col(X)
num_columns = num_col(X)
```

Перед разбивкой готовим входные признаки, закодированные при помощи `'LabelEncoder'`.

```python
X_le = X.copy()
le = LabelEncoder()
for col in cat_columns:
    X_le[col] = le.fit_transform(X_le[col])
```

Также отдельно готовим признаки для линейной регрессии, отмасштабированные с помощью `StandardScaler` и закодированные с помощью `OneHotEncoder`.

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_columns),
        ('cat', OneHotEncoder(
            drop='first', 
            handle_unknown='ignore', 
            sparse_output=False), 
         cat_columns)
    ], remainder='passthrough')

X_ohe = preprocessor.fit_transform(X)
```

Разбиваем признаки на тренировочные, валидационные и тестовые выборки.

```python
X_train_le, X_temp_le, y_train_le, y_temp_le = train_test_split(
    X_le, 
    y, 
    test_size=0.3, 
    random_state=RANDOM_STATE
)

X_val_le, X_test_le, y_val_le, y_test_le = train_test_split(
    X_temp_le, 
    y_temp_le, 
    test_size=0.5, 
    random_state=RANDOM_STATE
)
```

```python
X_train_ohe, X_temp_ohe, y_train_ohe, y_temp_ohe = train_test_split(
    X_ohe, 
    y, 
    test_size=0.3, 
    random_state=RANDOM_STATE
)

X_val_ohe, X_test_ohe, y_val_ohe, y_test_ohe = train_test_split(
    X_temp_ohe, 
    y_temp_ohe, 
    test_size=0.5, 
    random_state=RANDOM_STATE
)
```

Проверим размерность получившихся выборок.

```python
for sample_name in ['X_train_le', 'X_val_le', 'X_test_le']:
    sample = eval(sample_name)
    print(f'Доля {sample_name}: {sample.shape[0]/data.shape[0]:.0%}, '
          f'количество столбцов: {sample.shape[1]}')

print('\n')

for sample_name in ['y_train_le', 'y_val_le', 'y_test_le']:
    sample = eval(sample_name)
    print(f'Доля {sample_name}: {sample.shape[0]/data.shape[0]:.0%}')
```

```python
for sample_name in ['X_train_ohe', 'X_val_ohe', 'X_test_ohe']:
    sample = eval(sample_name)
    print(f'Доля {sample_name}: {sample.shape[0]/data.shape[0]:.0%}, '
          f'количество столбцов: {sample.shape[1]}')

print('\n')

for sample_name in ['y_train_ohe', 'y_val_ohe', 'y_test_ohe']:
    sample = eval(sample_name)
    print(f'Доля {sample_name}: {sample.shape[0]/data.shape[0]:.0%}')
```

Доли записей распределены корректно. После `OneHotEncoder` количество столбцов в выборках значительно увеличилось, что логично - каждое уникальное значение из категориальных признаков преобразовалось в отдельный столбец.

## Общий вывод

**Описание задачи**

В данной проектной работе мы помогли сервису по продаже автомобилей с пробегом «Не бит, не крашен» построить модель, которая поможет клиентам определить рыночную стоимость своего автомобиля. В процессе работы необходимо было учесть критерии, которые важны заказчику: качество предсказания, время обучения и время предсказания модели.

---

**Этапы работы**

<br>

*Загрузка и первичное изучение информации о данных:*

Всего нам было предоставлено 354 369 данных об автомобилях более 40 брендов. Анкеты были зарегистрированы в период с 2014 по 2016 годы.

<br>

*Предобработка данных:*

В результате предобработки было удалено 23.08% записей. В процессе мы:

- Удалили неинформативные признаки.

- Исправили неявные дубликаты и удалили явные (13.38%).

- Удалили 9.69% пропусков и аномальных значений, остальную часть мы заполнили модой и медианой по каждой модели автомобиля.

<br>

*Обучение моделей:*

Для задачи мы выбрали 4 модели: LinearRegression, RandomForestRegressor, LGBMRegressor, CatBoostRegressor. Наиболее оптимальные значения имеет модель CatBoost, ее время вычисления предсказания является самым быстрым, а также время ее обучения занимает второе место по скорости, при этом она показала наилучший уровень точности.

---

Заказчику рекомендуется использовать CatBoost для расчета рыночной стоимости автомобиля.
