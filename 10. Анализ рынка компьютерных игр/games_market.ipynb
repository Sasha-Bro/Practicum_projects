# Исследование данных о компьютерных играх

# Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Цель проекта

Выявить закономерности, определяющие успешность компьютерных игр, на основе исторических данных о продажах, оценках пользователей и экспертов, жанрах и платформах. Это позволит интернет-магазину прогнозировать популярность игр, планировать рекламные кампании и делать ставки на потенциально успешные продукты для кампании 2017 года.

Посмотрим, сколько лет составляет "срок жизни" платформ на рынке, а также через сколько лет наступает пик их популярности после первого появления на рынке.

```python
# Создаем пустой DataFrame для сохранения результатов
results = pd.DataFrame(columns=['platform', 'released', 'lifespan', 'years_to_peak'])

# Для каждой платформы
for platform in games_sales_pivot(data, 'total_sales', 'platform').index[:6]:
    platform_data = data[data['platform'] == platform]
    
    # Отображаем дату выхода платформы
    released = platform_data['year_of_release'].min()
    
    # Вычисляем срок жизни платформы
    lifespan = platform_data['year_of_release'].max() - released
    
    # Вычисляем, через сколько лет после даты выхода платформы наступает ее пик популярности
    peak_year = platform_data.groupby('year_of_release')['total_sales'].sum().idxmax()
    years_to_peak = peak_year - platform_data['year_of_release'].min()
    
    # Добавляем результаты в DataFrame
    results = results.append({'platform': platform, 'released': released, 'lifespan': lifespan, 'years_to_peak': years_to_peak}, ignore_index=True)

results
```

Среди платформ по количеству продаж за всё время лидируют PlayStation2, Xbox360, PlayStation3, Wii, NintendoDS и PlayStation.

### Влияние отзывов на продажи игр

Посмотрим, как влияют на продажи отзывы пользователей и критиков по каждой платформе.

Сперва получим список всех уникальных платформ, которые имеют оценки в столбцах `'critic_score'` и `'user_score'`.

```python
platforms = actual_data.dropna(subset=['critic_score', 'user_score'])['platform'].unique()
```

Построим диаграмму рассеяния между отзывами и продажами.

```python
# Создаем фигуру с двумя столбцами графиков (один для каждого типа корреляции)
fig, axs = plt.subplots(len(platforms), 2, figsize=(15, 5 * len(platforms)))

for i, platform in enumerate(platforms):
    platform_data = actual_data[actual_data['platform'] == platform]

    # Построение диаграммы рассеяния для 'critic_score' и 'total_sales'
    axs[i, 0].scatter(platform_data['critic_score'], platform_data['total_sales'])
    axs[i, 0].grid(True) 
    axs[i, 0].set_title(f'Корреляция между оценкой критиков и продажами для {platform}')
    axs[i, 0].set_xlabel('Оценка критиков')
    axs[i, 0].set_ylabel('Количество проданных копий, млн')

    # Построение диаграммы рассеяния для 'user_score' и 'total_sales'
    axs[i, 1].scatter(platform_data['user_score'], platform_data['total_sales'])
    axs[i, 1].grid(True)
    axs[i, 1].set_title(f'Корреляция между оценкой пользователей и продажами для {platform}')
    axs[i, 1].set_xlabel('Оценка пользователей')
    axs[i, 1].set_ylabel('Количество проданных копий, млн')

plt.tight_layout()
plt.show()
```

Посчитаем корреляцию между отзывами и продажами.

```python
# Создаем пустой DataFrame для хранения результатов и преобразуем данные в числовой формат
corr_data = pd.DataFrame(index=platforms, columns=['critic_score', 'user_score']).astype(float)

# Вычисляем корреляцию для каждой платформы
for platform in platforms:
    platform_data = actual_data[actual_data['platform'] == platform]
    corr_data.loc[platform, 'critic_score'] = platform_data['critic_score'].corr(platform_data['total_sales'])
    corr_data.loc[platform,'user_score'] = platform_data['user_score'].corr(platform_data['total_sales'])

# Отображаем данные с цветовой индикацией
plt.figure(figsize=(5, 8))
sns.heatmap(corr_data, annot=True, cmap='coolwarm');
```

Можно заметить, что в общем оценки критиков имеют большее значение, чем оценки пользователей. Когда как почти у всех платформ наблюдается средняя прямая корреляция между оценкой критиков и продажами, то между оценкой пользователей и продажами она очень слабая или не наблюдается вовсе. Чем выше оценка критиков, тем больше продаж. Покупатели склонны больше доверять отзывам критиков, чем отзывам других пользователей.

В списке выделяются платформы WiiU и Nintendo3DS, у которых значение оценки критиков и пользователей имеют почти одинаковое влияние на продажи. Также можно добавить, что у платформы PlayStationVita и у персональных компьютеров никакие оценки почти не имеют влияния на продажи.

Самая заметная зависимость продаж от оценки критиков видна у таких платформ, как:
- Xbox360 (53%), 
- PlayStation3 (45%), 
- XboxOne (43%), 
- PlayStation4 (40%).

## Проверка гипотез

Проверим гиспотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

Для проверки этих гипотез мы используем t-тест для двух независимых выборок. Этот тест используется для сравнения средних значений двух независимых групп, чтобы определить, есть ли статистически значимые различия между ними.

Зададим пороговое значение alpha в 5%.

```python
alpha = 0.05
```

### Пользовательские рейтинги платформ Xbox One и PC

**Гипотеза**: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

$H_0$: Средний пользовательский рейтинг платформы Xbox One **равен** среднему пользовательскому рейтингу PC.

$H_1$: Средний пользовательский рейтинг платформы Xbox One **не равен** среднему пользовательскому рейтингу PC.

```python
# выберем данные для Xbox One и PC
xone_data = actual_data[actual_data['platform'] == 'XOne']['user_score']
pc_data = actual_data[actual_data['platform'] == 'PC']['user_score']

# проведем t-тест
result_platform = stats.ttest_ind(xone_data, 
                                  pc_data, 
                                  equal_var=False, 
                                  nan_policy='omit')

print('Платформы XboxOne и PC:')
print('p-значение:', result_platform.pvalue)
if (result_platform.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, средние пользовательские рейтинги отличаются')
else:
    print('Не получилось отвергнуть нулевую гипотезу, средние пользовательские рейтинги не отличаются')
```

В результате t-теста мы не смогли отвергнуть нулевую гипотезу. Это означает, что на основании имеющихся данных мы не можем утверждать, что средние пользовательские рейтинги для платформ Xbox One и PC отличаются.

### Пользовательские рейтинги жанров Action и Sports

**Гипотеза**: Средние пользовательские рейтинги жанров Action и Sports разные.

$H_0$: Средний пользовательский рейтинг жанра Action **равен** среднему пользовательскому рейтингу жанра Sports.

$H_1$: Средний пользовательский рейтинг жанра Action **не равен** среднему пользовательскому рейтингу жанра Sports.

```python
# выберем данные для жанров Action и Sports
action_data = actual_data[actual_data['genre'] == 'Action']['user_score']
sports_data = actual_data[actual_data['genre'] == 'Sports']['user_score']

# проведем t-тест
result_genre = stats.ttest_ind(action_data, 
                               sports_data, 
                               equal_var=False, 
                               nan_policy='omit')

print('Жанры Action и Sports:')
print('p-значение:', result_genre.pvalue)
if (result_genre.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, средние пользовательские рейтинги отличаются')
else:
    print('Не получилось отвергнуть нулевую гипотезу, средние пользовательские рейтинги не отличаются')
```

В этом случае в результате t-теста мы отвергли нулевую гипотезу. Это означает, что на основании имеющихся данных мы можем утверждать, что существуют статистически значимые различия между средними пользовательскими рейтингами для жанров Action и Sports.

- Нулевая гипотеза отвергается. На основании имеющихся данных мы можем утверждать, что существуют статистически значимые различия между средними пользовательскими рейтингами для жанров Action и Sports.
